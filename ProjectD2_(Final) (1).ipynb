```python
for col in df.columns:
	# Check for null values presence
	if df[col].isnull().sum() == 0:
		continue

	# If the data type is categorical filling by mode.
	if df[col].dtype == object or df[col].dtype == bool:
		df[col] = df[col].fillna(df[col].mode()[0])

	# Else by mean
	else:
		df[col] = df[col].fillna(df[col].mean())

df.isnull().sum().sum()
```

```python
def classify_passenger_id(passenger_id):
    # Convert the passenger ID to a string and get the first 4 digits
    first_4_digits = str(passenger_id)[:4]

    # Count the occurrences of the first 4 digits in the DataFrame
    count = df["PassengerId"].astype(str).apply(lambda x: x.startswith(first_4_digits)).sum()

    # Return the classification based on the count
    if count > 1:
        return "family"
    else:
        return "single"

# Apply the classification function to the DataFrame
dftrain["Classification"] = dftrain["PassengerId"].apply(classify_passenger_id)

print(dftrain)
```
